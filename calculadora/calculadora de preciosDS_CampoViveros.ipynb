##resumen
El precio total de reservación en Airbnb se basa en el precio por noche establecida por el anfitrión, más cualquier otra tarifa o costo determinado por el anfitrión o Airbnb, de esta forma y sabiendo que los Airbnb tienen un mercado altamente competitivo, ya que compite con otros Airbnb, con la industria hotelera y aparta-estudios, sobre todo en una ciudad turística y grande como New York, establecer o predecir precios podría maximizar la ganancia y/u ocupación. El monitoreo predictivo de los precios de cualquier industria y/u organización tiene como objetivo pronosticar la adquisición de productos o servicios por parte de clientes y de esta forma ajustar la oferta para atraer a la demanda de esos servicios o productos. En el análisis de precios hay muchas formas de establecer el valor de un bien o servicio, generalmente se realiza un análisis de competencia o de mercado, análisis de datos históricos y análisis económicos por costos o por nivel de ganancia.

Los precios en los AIRBNB no escapan a estos análisis, establecer los precios nos ayudará a establecer ganancias hipotéticas y al ejecutar predicción de valores o precios, permitirá tener el control de estadía o de ocupación que conducirán a la maximización de ganancias; más que establecer un precio, en este proyecto deseamos realizar predicciones de precios para los Airbnb y en la ciudad de New York, ya que los precios varían según la temporada.

The total price of the reservation on Airbnb is based on the price per night established by the host, plus any other fees or costs determined by the host or Airbnb, taking into account that Airbnb has a highly competitive market, since it competes with other Airbnb, with the hotel industry and apart-studios (furnished apartments), especially in a large tourist city like New York, setting or predicting prices could maximize profit and/or occupancy. The predictive monitoring of prices of any industry and/or organization aims to forecast the acquisition of products or services by customers and thus adjust the offer to attract demand for those services or products. In the price analysis there are many ways to establish the value of a good or service, generally an analysis of competition or markets, analysis of historical data and economic analysis by costs or by level of profit are carried out.

Airbnb prices do not escape these analyses, establishing prices will help us to establish hypothetical profits and when executing prediction of values or prices it will allow control of stay or occupancy that will lead to the maximization of profits and/or benefits; We are not looking to simply establish a price, in this project we want to make price predictions for Airbnb in New York City, in this way multiple actors will benefit, from tenants, since they will be able to adjust their prices at appropriate times depending on the season, depending on of the District and will improve the level of occupancy of the airbnb, also to the Platform since it will be able to be much more competitive in front of the hotel industry and of rentals of houses and apartments, and finally to the client who takes the service, since he will obtain the best Airbnb available price in the specific region as of the requested date in New York

```python
from pandas.core.common import random_state
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_selector
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
import statsmodels.api as sm
from statsmodels.stats.diagnostic import linear_rainbow, het_breuschpagan
from statsmodels.stats.stattools import durbin_watson
from scipy.stats import shapiro
```

```python
X = New_airbnb[['habitacion','distrito','inm_reserv','cancelacion','barrio','number of reviews','reviews per month','calculated host listings count']]
Y = New_airbnb['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
y_train_prep = scaler.fit_transform(y_train.values.reshape(-1, 1))
y_test_prep=scaler.fit_transform(y_test.values.reshape(-1, 1))
print(x_train.shape,y_train_prep.shape)
print(x_test.shape, y_test_prep.shape)
x_train.isnull().sum()
```

```python
x_train_prep = scaler.fit_transform(x_train[['number of reviews','reviews per month','calculated host listings count']])
x_test_prep= scaler.transform(x_test[['number of reviews','reviews per month','calculated host listings count']])
x_test_scal = pd.DataFrame ( x_test_prep, columns=('number of reviews','reviews per month','calculated host listings count'),index=x_test.index)
x_train_scal = pd.DataFrame ( x_train_prep, columns= ('number of reviews','reviews per month','calculated host listings count'),index=x_train.index)

x_cat_trans= x_train[['habitacion','distrito','inm_reserv','cancelacion','barrio']]
x_cat_test= x_test[['habitacion','distrito','inm_reserv','cancelacion','barrio']]

x_train_prep = x_cat_trans.join(x_train_scal)
x_test_prep =x_cat_test.join(x_test_scal)

print(x_test_prep.isnull().sum())
print(x_train_prep.isnull().sum())
```

como se menciono anteriormente ya que las variables categoricas ya han sido escaladas o transformadas, en eset sentido solo se transformaran las variables numericas y se unen las variables numericas

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train_prep, y_train_prep)
y_pred = model.predict(x_test_prep)

print("Intercept:", model.intercept_)
print("Coeficiente:", list(zip(x_train_prep.columns, model.coef_.flatten(), )))
print("Coeficiente de determinación R^2:", model.score(x_train_prep, y_train))
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
print('root Mean squared error: %.2f' % np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
residuals = y_test_prep- y_pred
Residual = pd.DataFrame(residuals)
Residual.head()
```

##regresion polinomica

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=5)

X_train_poly = poly.fit_transform(x_train_prep)
X_test_poly = poly.transform(x_test_prep)

print(X_train_poly)
```

###### cross validation

```python
from sklearn.model_selection import TimeSeriesSplit

n_splits = 10
tscv = TimeSeriesSplit(n_splits=n_splits)

# Define los hiperparámetros que deseas probar
p = [0]
d = [0]
q = [0]
s = [12]  # Estacionalidad mensual

best_mse = float('inf')
best_params = None
```

en este caso los parametros se eligen segun las estadisticas y pruebas realizadas, no se aplicara diferencias entre la serie pues ya es estacionaria, no se aplica o introduce varibales autoregresivas, pero si se agrega o cambia la diferencia entre un modelo y otro y la media movil se pone en cero, porque en este caso solo se eligio un año en especifico para todos los datos

```python
for train_index, test_index in tscv.split(endog):
    endog_train, endog_test = endog[train_index], endog[test_index]
    exog_train, exog_test = exog.iloc[train_index], exog.iloc[test_index]

    for p_val in p:
        for d_val in d:
            for q_val in q:
                for s_val in s:
                    # Ajusta el modelo en los datos de entrenamiento con los hiperparámetros actuales
                    model = sm.tsa.SARIMAX(endog_train, exog=exog_train, order=(p_val, d_val, q_val), seasonal_order=(0, 0, 0, s_val))
                    results = model.fit()

                    # Calcula el MSE en los datos de prueba
                    forecast = results.forecast(steps=len(endog_test), exog=exog_test)
                    mse = mean_squared_error(endog_test, forecast)
```
