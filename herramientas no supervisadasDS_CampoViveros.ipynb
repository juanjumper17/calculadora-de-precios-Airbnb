#Prediccion de precios de AIRBNB en la Ciudad de New York 2003-2022

##resumen
El precio total de reservación en Airbnb se basa en el precio por noche establecida por el anfitrión, más cualquier otra tarifa o costo determinado por el anfitrión o Airbnb, de esta forma y sabiendo que los Airbnb tienen un mercado altamente competitivo, ya que compite con otros Airbnb, con la industria hotelera y aparta-estudios, sobre todo en una ciudad turística y grande como New York, establecer o predecir precios pudría maximizar la ganancia y/u ocupación. El monitoreo predictivo de los precios de cualquier industria y/u organización tiene como objetivo pronosticar la adquisición de productos o servicios por parte de clientes y de esta forma ajustar la oferta para atraer a la demanda de esos servicios o productos. En el análisis de precios hay muchas formas de establecer el valor de un bien o servicio, generalmente se realiza un análisis de competencia o de mercado, análisis de datos históricos y análisis económicos por costos o por nivel de ganancia.

Los precios en los AIRBNB no escapan a estos análisis, establecer los precios nos ayudará a establecer ganancias hipotéticas y al ejecutar predicción de valores o precios, permitirá tener el control de estadía o de ocupación que conducirán a la maximización de ganancias; más que establecer un precio, en este proyecto deseamos realizar predicciones de precios para los Airbnb y en la ciudad de New York, ya que los precios varían según la temporada.

The total price of the reservation on Airbnb is based on the price per night established by the host, plus any other fees or costs determined by the host or Airbnb, taking into account that Airbnb has a highly competitive market, since it competes with other Airbnb, with the hotel industry and apart-studios (furnished apartments), especially in a large tourist city like New York, setting or predicting prices could maximize profit and/or occupancy. The predictive monitoring of prices of any industry and/or organization aims to forecast the acquisition of products or services by customers and thus adjust the offer to attract demand for those services or products. In the price analysis there are many ways to establish the value of a good or service, generally an analysis of competition or markets, analysis of historical data and economic analysis by costs or by level of profit are carried out.

Airbnb prices do not escape these analyses, establishing prices will help us to establish hypothetical profits and when executing prediction of values or prices it will allow control of stay or occupancy that will lead to the maximization of profits and/or benefits; We are not looking to simply establish a price, in this project we want to make price predictions for Airbnb in New York City, in this way multiple actors will benefit, from tenants, since they will be able to adjust their prices at appropriate times depending on the season, depending on of the District and will improve the level of occupancy of the airbnb, also to the Platform since it will be able to be much more competitive in front of the hotel industry and of rentals of houses and apartments, and finally to the client who takes the service, since he will obtain the best Airbnb available price in the specific region as of the requested date in New York

```python
X = New_airbnb[['habitacion','distrito','inm_reserv','cancelacion','barrio','number of reviews','reviews per month','calculated host listings count']]
Y = New_airbnb['price']
```

```python
explained_variance = pd.DataFrame({
    'Component': ['PCA Component 1', 'PCA Component 2'],
    'Explained Variance': [pca.explained_variance_ratio_[0], pca.explained_variance_ratio_[1]]
})
print(explained_variance)
```

al realizar el analisi de componentes principales se puede observar que el òrcentaje explicado de la varianza es de en el primer componente el 99%, esto es positivo, debido a que en 1 solo componente se encuentra reducida todas las variables.

#### UMAP

el UMAP o La aproximación y proyección de colector uniforme (UMAP), es un metodo no lineal para reducir la dimensiones, una de sus caracteristicas o supuestos es que la data debe estar uniformemente distribuida, como en este caso concreto

```python
umap_result = umap.UMAP(n_components=2).fit_transform(X_df)
umap_result
```

UMAP esta orientado a la interconexion de los datos, no a capturar tanto la mayor varianza que explica los datos como el PCA, es decir, está más orientado a capturar la interconexión entre los puntos de datos que la varianza.

```python
fig, ax= plt.subplots(figsize= (15,10))
ax.scatter(umap_result[:,0],  umap_result[:,1],c='red',alpha=0.5)
plt.title('UMAP (Reduccion dimensional)',fontsize=16)
ax.set_xlabel('Componente 1',fontsize=16)
ax.set_ylabel('Componente 2',fontsize=16)
plt.show()
```

comparacion entre los dos metodos de reduccion de dimensiones, como se puede observar y ya se ha mencionado, el UMAP esta adaptado a realizar una reduccion no lineal del dataset ademas que se ajusta a las bases de datos uniformemente distribuidas.
el PCA trata de captura la maxima varianza explicada ademas de ser un metodo lineal los resultados. finalemnete se tiene la grafica conjunta

####Kmeans

El algoritmo K-means tiene como objetivo minimisar la inercia (VER MAS EN: https://scikit-learn.org/stable/modules/clustering.html#k-means) , o lo que es lo mismo la diferencia al cuadradp de el dato respecto a la media del grupo
La inercia supone que los cúmulos son convexos e isotrópicos(que es igual en todas las direcciones), lo que no siempre es así.
Responde mal a racimos alargados o variedades con formas irregulares, como esn este caso
La inercia no es una métrica normalizada: Pen espacios de dimensiones muy altas, las distancias euclidianas tienden a inflarse, sin embargo la forma de aliviar esto es utilizar el analisis de componenetes principales

```python
k_values = range(1, 10)
kmeans_wcss = [] #suma de la diferencia entre los datos y la media al cuadrado
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=1).fit(X)
    # Compute the WCSS for KMeans
    kmeans_wcss.append(kmeans.inertia_)
    plt.figure(figsize=(10, 5))
plt.plot(k_values, kmeans_wcss, '-x')
plt.xlabel('numero de clusters, k')
plt.ylabel('WCSS (Inercia)')
plt.title('metodo del codo')
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 4), dpi=120)
km = KMeans(n_clusters=2, random_state=0)
y_km = km.fit_predict(X_pca)
ax[0].scatter(X_pca[:, 0], X_pca[:, 1], c=y_km)
ax[0].set_title('KMeans PCA')

y_km = km.fit_predict(umap_result)
ax[1].scatter(umap_result[:, 0], umap_result[:, 1], c=y_km)
ax[1].set_title('KMeans UMAP')

```

#### afinidad de propagacion (ap)

affinity_propagation = cluster.AffinityPropagation()

(AP) es un algoritmo que trabaja con la matriz de similitud, en vez de con la distancia euclídea,Esto hace que los puntos que están cerca entre sí, tienen una alta similitud y los más alejados tienen una baja similitud, es decir se parecen poco.

La Matriz de Responsabilidad: es la matriz encargada de determinar como de responsable es cada punto de nuestro conjunto de datos, es decir la importancia que tiene.

La Matriz de Disponibilidad: La matriz encargada de determinar la cantidad de puntos que tiene un punto al rededor, es decir sus vecinos.

Al contrario que otros algoritmos de clustering, a este no es necesario decir el número de clusters que se quieren


#####hiperparametros

dumping: es el factor que controla el número de clusters, va de 0.5 a 1. Cuanto mayor es el dumping, menos clusters habrá.

```python
afinidad = AffinityPropagation(damping=0.5,random_state=0, max_iter = 20)
y_ap= afinidad.fit_predict(X_pca)
```

```python
ax[0].scatter(X_pca[:, 0], X_pca[:, 1], c=y_ap)
ax[0].set_title('KMeans PCA')

y_ap= afinidad.fit_predict(umap_result)
ax[0].scatter(umap_result[:, 0], umap_result[:, 1], c=y_ap)
ax[0].set_title('KMeans PCA')
```

en este caso el cluster encontro cuatro grupos en el pca, debido a la separacion en el dataset, y tres en el umap seguramente debido a la explicadion dada antes sobre las distancias en el dataset, ya que no se pone el numero de clusters sin embargo si se puede controlar por el damping o la amortiguacion sin embargo, a mayor iteraciones mayor consumo de cpu y memoria ram.
