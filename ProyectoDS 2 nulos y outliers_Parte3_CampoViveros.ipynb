#Prediccion de precios de AIRBNB en la Ciudad de New York 2003-2022

##resumen
El precio total de reservación en Airbnb se basa en el precio por noche establecida por el anfitrión, más cualquier otra tarifa o costo determinado por el anfitrión o Airbnb, de esta forma y sabiendo que los Airbnb tienen un mercado altamente competitivo, ya que compite con otros Airbnb, con la industria hotelera y aparta-estudios, sobre todo en una ciudad turística y grande como New York, establecer o predecir precios pudría maximizar la ganancia y/u ocupación. El monitoreo predictivo de los precios de cualquier industria y/u organización tiene como objetivo pronosticar la adquisición de productos o servicios por parte de clientes y de esta forma ajustar la oferta para atraer a la demanda de esos servicios o productos. En el análisis de precios hay muchas formas de establecer el valor de un bien o servicio, generalmente se realiza un análisis de competencia o de mercado, análisis de datos históricos y análisis económicos por costos o por nivel de ganancia. 

Los precios en los AIRBNB no escapan a estos análisis, establecer los precios nos ayudará a establecer ganancias hipotéticas y al ejecutar predicción de valores o precios, permitirá tener el control de estadía o de ocupación que conducirán a la maximización de ganancias; más que establecer un precio, en este proyecto deseamos realizar predicciones de precios para los Airbnb y en la ciudad de New York, ya que los precios varían según la temporada.

The total price of the reservation on Airbnb is based on the price per night established by the host, plus any other fees or costs determined by the host or Airbnb, taking into account that Airbnb has a highly competitive market, since it competes with other Airbnb, with the hotel industry and apart-studios (furnished apartments), especially in a large tourist city like New York, setting or predicting prices could maximize profit and/or occupancy. The predictive monitoring of prices of any industry and/or organization aims to forecast the acquisition of products or services by customers and thus adjust the offer to attract demand for those services or products. In the price analysis there are many ways to establish the value of a good or service, generally an analysis of competition or markets, analysis of historical data and economic analysis by costs or by level of profit are carried out.

Airbnb prices do not escape these analyses, establishing prices will help us to establish hypothetical profits and when executing prediction of values or prices it will allow control of stay or occupancy that will lead to the maximization of profits and/or benefits; We are not looking to simply establish a price, in this project we want to make price predictions for Airbnb in New York City, in this way multiple actors will benefit, from tenants, since they will be able to adjust their prices at appropriate times depending on the season, depending on of the District and will improve the level of occupancy of the airbnb, also to the Platform since it will be able to be much more competitive in front of the hotel industry and of rentals of houses and apartments, and finally to the client who takes the service, since he will obtain the best Airbnb available price in the specific region as of the requested date in New York
